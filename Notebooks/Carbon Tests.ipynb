```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Results = pd.read_csv("../Scripts/Result.csv")
print Results.columns
print Results.shape
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = plt.axes([0.1,0.5,0.8,0.4])
ax2 = plt.axes([0.1,0.1,0.8,0.4], sharex=ax1)

Resid = np.array(Results.CFE_EST - Results.CFE_COR)

ax1.errorbar(Results.CFE_COR, Results.CFE_EST, xerr=Results.CFE_ERR, yerr=Results.CFE_Sigma,
                fmt="o", markersize=2.0, alpha=0.7, color="black")

ax1.plot(np.linspace(-1.5,3, 10), np.linspace(-1.5,3, 10), linestyle="--")
# ax1.scatter()


ax2.errorbar(Results.CFE_COR, Results.CFE_EST - Results.CFE_COR,
             yerr=Results.CFE_Sigma, xerr=Results.CFE_ERR,
             fmt="o", markersize=2.0, alpha=0.7, color="black")
ax2.plot(np.linspace(-1.5, 3, 50), np.zeros(50), linestyle="--")

ax2.set_ylabel("Residual", fontsize=14, family = "serif", fontname = "Courier")
ax1.set_ylabel("[C/Fe] Estimate", fontsize=14, family = "serif", fontname = "Courier")
ax2.set_xlabel("[C/Fe] Biweight", fontsize=14, family = "serif", fontname = "Courier")
```
